```python
# --- Load data ---
from models.model_utils import params
persons = [0.01, 0.3, 0.7, 1.5]
initial_model = LinearModel_Estimator(**params)
print(f"Initial Q matrix: {initial_model.Q}")


'''STOPPED HERE - NEED TO FIX SO THE CALIBRATORS AND THE CALIBRATED MODEL WILL BE SAVED IN A LIST FOR EACH PERSON'''
''' EACH METHOD WILL HAVE 2 LISTS: ONE FOR THE CALIBRATORS AND ONE FOR THE CALIBRATED MODELS'''ArithmeticError

calibrators = []
calibrated_models_B_CUC = []
metrics_qc_array = []

# --
for person in persons:
    path_to_load = f"../observations/observations_random_zeros_and_actions_for_Q{person}_steps_10000.npz"
    data = np.load(path_to_load, allow_pickle=True)
    masked_observations = mask_observations(data['observations'])
    actions = data['actions']

    # --- build models ---
    to_print = f"Target (true) Q*: {person}"
    print(to_print)
    
    cal_qc = BCUC_Calibrator(
        initial_model=initial_model,
        alpha=0.32,          # μ±σ target; keeps Q identification unbiased if Gaussian
    )
    model_qc = cal_qc.calibrate_model(masked_observations, actions)
    calibrated_models.append(model_qc)
    calibrators.append(cal_qc)

    cal_dq = DirectQBCUCCalibrator(
        initial_model=initial_model,
        gamma=0.5,
        lambd=0.01,
        update_after_every=15
    )
    model_dq = cal_dq.calibrate_model(masked_observations, actions)
    calibrated_models.append(model_dq)
    calibrators.append(cal_dq)

    cal_nt = NLLThresholdCalibrator(
        initial_model=initial_model,
        gamma=0.5,
        k=50,
        alpha=0.32,
        threshold_nll=1.0,
        lambda_=0.01,
        update_after_every=15
    )
    model_nt = cal_nt.calibrate_model(masked_observations, actions)
    calibrated_models.append(model_nt)
    calibrators.append(cal_nt)

    # --- evaluate metrics ---
    metrics_qc = cal_qc.get_metrics(person)
    metrics.append(metrics_qc)
    metrics_dq = cal_dq.get_metrics(person)
    metrics.append(metrics_dq)
    metrics_nt = cal_nt.get_metrics(person)
    metrics.append(metrics_nt)
    
    # --- print results ---
    print("[BCUC quantile_coupled]")
    print("metrics:", metrics_qc)
    print("final Q:", model_qc.Q)
    print()
    print("[Direct Q Calibrator]")
    print("metrics:", metrics_dq)
    print("final Q:", model_dq.Q)
    print()
    print("[NLL Threshold Calibrator]")
    print("metrics:", metrics_nt)
    print("final Q:", model_nt.Q)
    print()
```
